Q2. State and Explain the features of MongoDB.

```python
#another example for find() and find_one()

from pymongo import MongoClient

# Connect to the MongoDB server
client = MongoClient('mongodb://localhost:27017/')

# Access the database
db = client['mydatabase']

# Access the collection
collection = db['mycollection']

# Insert a document
document = { 'name': 'John', 'age': 30 }
inserted_doc = collection.insert_one(document)
inserted_id = inserted_doc.inserted_id

# Retrieve and print the inserted record using find_one()
result = collection.find_one({ '_id': inserted_id })
print(result)

# Retrieve and print the inserted record using find()
results = collection.find({ '_id': inserted_id })
for doc in results:
    print(doc)
```

Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.

To sort the documents, we use the sort() method on the find() method's result.    
The sort() method modifies the ordering of the result cursor and does not change the actual order of the documents in the collection.

#Example as below--

from pymongo import MongoClient                              

#Connect to the MongoDB server                                      
client = MongoClient('mongodb://localhost:27017/')                       

#Access the database                         
db = client['mydatabase']

#Access the collection                      
collection = db['mycollection']

#Insert some documents                                                   
collection.insert_many([                              
    { 'name': 'John', 'age': 30 },                       
    { 'name': 'Alice', 'age': 25 },                     
    { 'name': 'Bob', 'age': 35 }                 
])

#Sort the documents by name in ascending order                       
result = collection.find().sort('name', 1)                             
for doc in result:                                
    print(doc)                                       

#Sort the documents by age in descending order                                 
result = collection.find().sort('age', -1)                        
for doc in result:                                    
    print(doc)                 

Q7. Explain why delete_one(), delete_many(), and drop() is used.

The delete_one(), delete_many(), and drop() methods are used for deleting documents or collections from a database.                        

delete_one(filter):                                  

This method is used to delete a single document that matches the specified filter criteria.
It takes a filter as a parameter, which defines the criteria for matching the document to be deleted.
If multiple documents match the filter, only the first one encountered will be deleted.
If no document matches the filter, no deletion occurs.
This method is useful when you want to delete a specific document based on certain conditions.
delete_many(filter):

This method is used to delete multiple documents that match the specified filter criteria.
It takes a filter as a parameter, similar to delete_one().
All documents that match the filter will be deleted.
If no document matches the filter, no deletion occurs.
This method is useful when you want to delete multiple documents based on certain conditions.
drop():

This method is used to delete an entire collection from the database.
It does not take any parameters and directly drops the entire collection.
All documents, indexes, and metadata associated with the collection will be permanently removed.
This method is useful when you want to completely remove a collection and all its data.
