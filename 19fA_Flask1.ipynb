Q1. What is Flask Framework? What are the advantages of Flask Framework?

-Flask is a micro web framework written in Python. It is designed to be lightweight and simple, providing a flexible environment for building web applications. 

**Advantages:
1. Lightweight and Minimalistic: Flask has a minimalistic design, focusing on the core functionality required for web development           
2. Easy to Get Started: Flask has a simple and intuitive syntax, making it easy for developers to get started with building web applications.          
3. Flexibility: Flask provides a high level of flexibility, allowing developers to choose the components they need for their project. It supports extensions and integrates well with other libraries, giving developers the freedom to customize their application according to their requirements.      
4. Extensibility: Flask offers a wide range of extensions and libraries that can be easily integrated into the framework.               
5. Scalability: Flask is scalable and can handle projects of varying sizes. It allows developers to scale their applications as needed, ensuring smooth performance even with increased user traffic.

Q3. What is App routing in Flask? Why do we use app routes?

-App routing is a process of mapping URLs to python functions in flask application

-App routes are used in flask for following reasons--
1. URL Mapping: App routes allow you to define the URLs or routes that users can access to interact with different parts of your application.          
2. Handling Different HTTP Methods: App routes can be configured to handle different HTTP methods such as GET, POST, PUT, DELETE, etc.               
3. Dynamic URL Parameters: App routes support dynamic URL parameters, which allow you to extract values from the URL and pass them as arguments to your view functions.                       
4. Modular Structure: By using app routes, you can organize your Flask application into modular views. Each view can handle specific routes and related functionality              
5. Decorator-based Syntax: Flask uses decorators to define app routes, which provides a clean and concise syntax for associating routes with view functions.

Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

-In Flask, the url_for() function is used for URL building. It helps generate URLs for routes defined in the application, making it easier to handle dynamic URLs and ensure proper URL structure.

--code as below

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')                          
def home():                                                
    return 'Welcome to the Home Page'                                   

@app.route('/about')                                            
def about():                                          
    return 'This is the About Page'                               

@app.route('/contact')                                    
def contact():                                                      
    return 'Contact us at example@example.com'                            

if __name__ == '__main__':                                      
    with app.test_request_context():                                 
        # Generate URLs using the url_for() function                          
        home_url = url_for('home')                        
        about_url = url_for('about')                          
        contact_url = url_for('contact')                       

        # Print the generated URLs                              
        print('Home URL:', home_url) 
        print('About URL:', about_url)
        print('Contact URL:', contact_url)
